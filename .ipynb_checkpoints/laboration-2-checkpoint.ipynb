### Comparing sequences

We start by describing how we can compare sequences.  

All of our examples will be sequences that use only the letters $\{A,C,G,T\}$. These letters represent the four nucleotide bases of a DNA strand: adenine, cytosine, guanine and thymine. Genomes are given as very long sequences of these letters. Under the assumption that similar sequences can have similar functionality, it turns out to be very useful to be able to compare fragments of genome sequences and identifying fragments that are similar. One way of doing this is by introducing gaps in one or both of the sequences to align similar fragments.

####  Sequence alignment
In **Exercise 1** we could have considered the second sequence as having a gap at the end, then both sequences would have been of the same length.

Aligning two sequences $X$ and $Y$ is about inserting gaps to one or both so that both extended sequences have the same length. 

Two sequences can be aligned in different ways (you saw this in **Exercise 1**). Some alignments are better than others to reveal how similar two sequences are. So, how can we measure this quality of the alignment? We can introduce so called *penalties* for each column where we do not have the same letter. Both gaps and mismatches generate  a penalty and they can all be different. 

Aligning two sequences $X$ and $Y$ is about inserting gaps to one or both so that both extended sequences have the same length and the total penalty is as small as possible. This total smallest penalty is called the Needleman‚ÄìWunsch (N-W) score of the sequences. This score is used to say how similar two sequences are.

#### The recursive version
We right down the solution arrived at in the videos. 

Given two sequences $X$ of length $m$ and $Y$ of length $n$, let $P_{i,j}$ be the N-W score of aligning $X_i$ and $Y_j$. We assume that the penalty for a gap is $\alpha_{\text{gap}}$ and the penalty for two letters $x$ and $y$ is $\alpha_{xy}$ (0 if the letters are identical).

Here is how to calculate the N-W score for $X$ and $Y$, that is $P_{m,n}$:

$P_{m,0} = m \cdot \alpha_{\text{gap}}$

$P_{0,n} = n \cdot \alpha_{\text{gap}}$

$P_{m,n} = \text{min}\{P_{m-1,n-1} + \alpha_{x_{m}y_{n}}, P_{m-1,n} + \alpha_{\text{gap}}, P_{m,n-1} + \alpha_{\text{gap}}\}$

**Important:** In the equations we use indexing of the letters in the sequence starting with 1 (as in the book and the videos): $X = x_1x_2\ldots x_m$ and $Y = y_1y_2 \ldots y_n$.

####  Exercise 4
1. Write a Python function that takes as arguments:

    - two sequences,

    - one penalty for mismatching letters

   and returns the N-W score of the two sequences. *The function should be an implementation of the recursive equations given above*. Use strings for the sequences and remember that the first position in a string has index 0!

2. Test your program with the strings from exercise 1, gap penalty of 1 and mismatch penalty of 2.

````python
# SOLUTION TO EXERCISE 4.1
def nw_align_recursive(x, y, alpha_gap, alpha_xy):
    
    if(len(x) == 0):
        return len(y)*alpha_gap
    
    if(len(y) == 0):
        return len(x)*alpha_gap
    
    if(x[-1] == y[-1]):
        alpha_xiyj = 0
    else:
        alpha_xiyj = alpha_xy      
        
    penalty = min(nw_align_recursive(x[:-1], y[:-1], alpha_gap, alpha_xy) + alpha_xiyj, nw_align_recursive(x[:-1], y, alpha_gap, alpha_xy) + alpha_gap, nw_align_recursive(x, y[:-1], alpha_gap, alpha_xy) + alpha_gap)
    print(x)
    print(y)
    return penalty
    
````

#### The dynamic programming version

We write down the solution presented in the videos.

*Input:* Sequences $X = x_1x_2\ldots x_m$ and $Y = y_1y_2\ldots y_n$. A penalty for gaps $\alpha_{\text{gap}} \geq 0$. A penalty $\alpha_{xy}$ for letters $x$ and $y$.

*Output:* The N-W score for $X$ and $Y$.

```
A := (m + 1) x (n + 1)   // a two dimensional array indexed from 0

for i := 0 to m do
    A[i][0] = i * alpha_gap

for j := 0 to n do
    A[0][j] = j * alpha_gap

for i := 1 to m do
     for j := 1 to n do
         A[i][j] := min of 
                      A[i-1][j-1] + alpha_xi_yj
                      A[i-1][j] + alpha_gap 
                      A[i][j-1] + alpha_gap 
return A[m][n]                      


```

   that *joins* all the strings in the list using ```''``` (the empty string) between each of the letters. For example,

````python
# SOLUTION TO EXERCISE 5.1

def nw_align_dynamic(x, y, alpha_gap, alpha_xy):
    
    print(len(x), len(y))
    
    penalty = [None]*(len(y)+1)
    for pos in range(len(y)+1):
        penalty[pos] = [0]*(len(x)+1)

    for i in range(len(x)+1):
        penalty[0][i] = i * alpha_gap
        pass

    for i in range(len(y)+1):
        penalty[i][0] = i * alpha_gap
        pass
    
    
    for xi in range(1, len(x)+1):
        for yj in range(1, len(y)+1):
            if(x[xi-1] == y[yj-1]):
                alpha = 0
            else:
                alpha = alpha_xy
            penalty[yj][xi] = min(penalty[yj-1][xi-1] + alpha, penalty[yj][xi-1] + alpha_gap, penalty[yj-1][xi] + alpha_gap)
            
    return penalty[-1][-1]
````

````python
# SOLUTION TO EXERCISE 5.2
nw_align_dynamic('AGGGCT', 'AGGCA', 1, 2)
````

````python
# SOLUTION TO EXERCISE 5.3
import random
import time

def test_time_nw():
    sizes = [1000, 2000, 4000, 8000, 16000]
    result = []
    for i in sizes:
        X = ''.join(random.choices('ATCG', k = i))
        Y = ''.join(random.choices('GCTA', k = i))
        
        start1 = time.process_time()
        nw_align_recursive(X, Y, 1, 2)
        end1 = time.process_time()
        
        start2 = time.process_time()
        nw_align_dynamic(X, Y, 1, 2)
        end2 = time.process_time()
        result.append((i, end1-start1, end2- start2))
    return result
        
test_time_nw()
````
