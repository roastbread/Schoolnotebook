```python
#Please implement task 1a here
energy_data = energy_data.drop('date', axis=1)
```

```python
#Please implement task 1b here
from sklearn.preprocessing import MinMaxScaler
usage_level = energy_data['Usage_level']
energy_data = energy_data.drop('Usage_level', axis=1)
norm_data = (energy_data - energy_data.min(axis=0)) / (energy_data.max(axis=0) - energy_data.min(axis=0))
```

```python
#Please implement task 1c here
labels = usage_level
split = int(0.7 * len(norm_data))

train_data = norm_data[:split]
test_data= norm_data[split:]
train_labels= labels[:split]
test_labels=labels[split:]
```

```python
#Please implement task 2a here
from pyod.models.pca import PCA
contamination = sum(labels)/len(labels)
pca = PCA(contamination=contamination)
X_pca = pca.fit(norm_data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
#Five looks good
```

```python
#Please implement task 2b here
pca = PCA(n_components=5, contamination=contamination)
pca.fit(train_data)
train_scores = pca.predict_proba(train_data)
test_scores = pca.predict_proba(test_data)
```

```python
#Please implement task 2c here
import sklearn.metrics as met

def true_positive(ground_truth, predicted, treshold):
    true = predicted[ground_truth==1]
    tp = true[true<treshold]
    return len(tp)

def false_positive(ground_truth, predicted, treshold):
    false = predicted[ground_truth==0]
    fp = false[false<treshold]
    return len(fp)

def false_negative(ground_truth, predicted, treshold):
    false = predicted[ground_truth==0]
    fn = false[false>treshold]
    return len(fn)

def pr_curve(ground_truth, predicted, range_):
    precision = []
    recall = []
    for i in range_:
        tp = true_positive(ground_truth, predicted, i)
        fp = false_positive(ground_truth, predicted, i)
        fn = false_negative(ground_truth, predicted, i)
        try:
            pr = (tp/(tp + fp))
            rc = (tp/(tp + fn))
            precision.append(pr)
            recall.append(rc)
        except:
            continue
    return precision, recall

tp = true_positive(test_labels, test_scores[:,0], 0.5)
fp = false_positive(test_labels, test_scores[:,0], 0.5)
fn = false_negative(test_labels, test_scores[:,0], 0.5)
print(tp/(tp + fp), 'precision')
print(tp/(tp + fn), 'recall')
x = np.linspace(0.2,0.7,100)
test_pr = pr_curve(test_labels, test_scores[:,0], x)
train_pr = pr_curve(train_labels, train_scores[:,0], x)

train_precision = met.precision_score(train_labels, np.round(train_scores[:,0]), pos_label = 0)
test_precision = met.precision_score(test_labels, np.round(test_scores[:,0]), pos_label = 0)
train_recall = met.recall_score(train_labels, np.round(train_scores[:,0]), pos_label = 0)
test_recall = met.recall_score(test_labels, np.round(test_scores[:,0]), pos_label = 0)


train_pc_curve = met.precision_recall_curve(train_labels, train_scores[:,0], pos_label = 0)
test_pc_curve = met.precision_recall_curve(test_labels, test_scores[:,0], pos_label = 0)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9,4))
ax1.plot(train_pc_curve[1], train_pc_curve[0])
ax2.plot(test_pc_curve[1], test_pc_curve[0])
ax1.set_title('Train precision-recall curve')
ax2.set_title('Test precision-recall curve')
ax1.set_xlabel('Recall')
ax1.set_ylabel('Precision')
ax2.set_xlabel('Recall')
print(train_precision, 'train precision')
print(train_recall, 'train recall')
print(test_precision, 'test precision')
print(test_recall, 'test_recall')
print(met.auc(train_pc_curve[1], train_pc_curve[0]), 'train_auc')
print(met.auc(test_pc_curve[1], test_pc_curve[0]), 'test_auc')

#As can be expected the training dataset performs better than the test data
```
