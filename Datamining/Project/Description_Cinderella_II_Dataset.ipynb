### Weather Variables

Onb, stands for wind variables which are from Onboard data (from the vessel sensors)  
cpr, stands for wind variables which are from Climate Data Store - Copernicus data, external source [https://cds.climate.copernicus.eu](https://cds.climate.copernicus.eu)  
sg, stands for wind variables which are from Stormglass Data, external source [https://stormglass.io](https://stormglass.io)  
No subscript, means this weather variable has a unique name and no need to distinguish it from other variables, the variable either from Stormglass or Copernicus sources.

```python
df[['windDirection_Onb','windSpeed_Onb','windSpeed_cpr','windDirection_cpr','windSpeed_sg','windDirection_sg',
    'waveHeight', 'waveDirection','currentSpeed', 'currentDirection']].describe()
```

### Example of Data Visualization
Plot all the vessel's routes and save the output plot in the the created output folder

```python
# Create a scatter plot of latitude and longitude to visualize the ferry's path
fig = plt.figure(figsize=(10,7))
plt.scatter(df['longitude'], df['latitude'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid()

plt.savefig(Output_folder+'/Scatter_dforig_lat_lon_paths'+'.png', dpi=500)  

plt.show()
```
