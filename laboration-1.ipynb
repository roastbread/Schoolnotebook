### Partitioning an array

We start by showing Python code to reorder the elements of an array so that three elements get sorted: the first element, the element in the middle position and the last element. 

In order to sort the three elements we need to identify the middle position and we need to be able to swap elements. 


````python
def swap(a,i,j):
    a[i],a[j] = a[j],a[i]
    
def sort_three(a):
    n = len(a) - 1
    mid = n // 2

    
    if a[mid] < a[0]:
        swap(a, 0, mid)
    if a[n] < a[0]:
        swap(a, 0, n)
    if a[n] < a[mid]:
        swap(a, mid, n)
````

Now that you know how to sort the three elements, you will see code that rearranges the elements of a slice in an array so that all the elements *smaller or equal* to the middle one (called the pivot) appear before the pivot and all elements *larger or equal* to the pivot appear after the pivot. This is often called partition.

The function ```partition``` re-arranges the array **and** returns the position of the pivot in the array.

````python
# Before start we know that a[lo] <= a[mid] <= a[hi]

def partition(a, lo, hi):
    
    # Place the pivot out of the way, in position hi - 1 
    # The element in position hi is >= pivot so we keep it there! 
    mid = (lo + hi)//2
    swap(a, mid, hi - 1)
    
    # initialize the indices to walk through the array from low to high and from hight to low
    # i starts from the left
    i = lo 
    
    # j starts from the right
    j = hi - 1 
    
    pivot = a[j]
    
    # the algorithm that Tim explains
    while True:
        while True:
            i += 1
            if a[i] >= pivot: break
        # here you know that a[i] >= pivot!
           
        while True:
            j -= 1
            if a[j] <= pivot: break
        # here you know that a[i] >= pivot and a[j] <= pivot
            
        if i >= j: break # you are done if the indices have crossed each other!
            
        # here you know that a[i] >= pivot and a[j] <= pivot and i < j!
        swap(a, i, j)

    # put the pivot in its place and return the position of the pivot    
    swap(a, i, hi - 1)
    return i
````

Finally, here is a program that implements the algorithm. I hope you recognize the programs we used in the exercises 1, 2 and 3. 
    
In Python we usually give names starting with underscore to auxiliary functions.
    
</div>

````python
def quicksort(a):
    _qsort(a,0,len(a)-1)
    
def _qsort(a, lo, hi):
    # For small arrays insertion sort is to be prefered: no recursion.
    # The implementation of sorted in Python uses insertion sort 
    # for arrays of length less than 64. So we can use that. 
    # The base case is then
    #     if lo + 64 > hi: a[lo:hi+1] = sorted(a[lo:hi+1]) 

    # but, at least three elements for the base case 
    # (think of partition, needs 3 elements to work properly)
    
    if lo + 64 > hi: a[lo:hi+1] = sorted(a[lo:hi+1]) 
    else: 
        _pivot(a, lo, hi)
        i = _partition(a, lo, hi)
        _qsort(a, lo, i - 1)
        _qsort(a, i + 1, hi)
        
    
# Now the functions we studied in the previous exercises.
# Here with names as auxiliary functions (starting with _) 
# Also, we use the name _pivot instead of sort_three because it could be 
# some other function than median of three!
# For example, you could pick a random value in the slice as pivot!

def _swap(a,i,j):
    a[i], a[j] = a[j], a[i]
    
def _pivot(a, lo, hi):
    mid = (lo + hi) // 2
    # sort lo, mid, hi:
    if a[mid] < a[lo]:
        _swap(a, lo, mid)
    if a[hi] < a[lo]:
        _swap(a, lo, hi)
    if a[hi] < a[mid]:
        _swap(a, mid, hi)

def _partition(a, lo, hi):
    # place the pivot out of the way, in position hi -1 
    mid = (lo + hi)//2
    _swap(a, mid, hi - 1)
    
    # initialize the indices to walk through the array from low to high and from hight to low
    i = lo
    j = hi - 1
    pivot = a[j]
    
    # the algorithm that Tim explains
    while True:
        while True:
            i += 1
            if a[i] >= pivot: break
           
        while True:
            j -= 1
            if a[j] <= pivot: break
            
        if i >= j: break
        swap(a, i, j)

    # put the pivot in place and return the position of the pivot    
    _swap(a, i, hi - 1)
    return i
````

The result of the function ```tabulate_T_n(start, times)``` is a list of pairs. Each pair has a size and the average time for that size.

Remember that we are interested in big sizes (one of the guiding assumptions for doing asymptotic analysis). We can think of this list as a table:

|    N      | T(N)  |
|:---------:|:-----:|
| start     | $t_0$ |
| 2 * start | $t_1$ |
| 4 * start | $t_2$ |
| 8 * start | $t_3$ |
| $\ldots $ | $\ldots$|
| $2^9$ * start|  $t_9$ |

You do not need to print it as a table: just generate a list of pairs ```[(start, T(start)), (2*start,T(2*start)), ..., (size, T(size)]```. And be prepared to wait!

<hr style="border:1px solid silver"> </hr>


````python
# SOLUTION TO EXERCISE 4
import time
def tabulate_T_n(start, times):
    table = []
    for power in range(10):
        avg = 0
        for test in range(10):
            test_case = list(range(start * 2**power))
            random.shuffle(test_case)
            t1 = time.process_time()
            quicksort(test_case)
            t2 = time.process_time() -t1
            avg += t2
        avg = avg/10
        table.append((len(test_case), avg))
        print(avg)
    return table
table = tabulate_T_n(800, 10)
print(table)
````
